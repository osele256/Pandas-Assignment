```python
#Total Sales in North America for the 'Action' Genre
action_df = df[df['Genre'] == 'Action']
total_sales_NA = action_df['NA_Sales'].sum()
print("Total Sales in North America for the Action genre: ",total_sales_NA)

```

```python
#Average sales in Europe for the Shooter genre

Shooter_df = df[df['Genre'] == 'Shooter']
total_sales_EU = Shooter_df['EU_Sales'].mean()
print("Average Sales in Europe for the Shooter genre: ", total_sales_EU)

```

```python
#Game with the highest sales in japan in the sports genre
sports_df = df[df['Genre'] == 'Sports'] # filter out Dataframe for 'Sports' genre

  #game with highest sales
highest_sales_game = sports_df.loc[sports_df['JP_Sales'].idxmax()]
  #display result
print("The game with the Highest Sales in japan in the sports genre: ", highest_sales_game)
```
